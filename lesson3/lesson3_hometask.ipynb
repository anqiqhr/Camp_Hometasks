```python
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.float_format', '{:.3f}'.format)
pd.set_option('display.max_columns', None)
import os
cwd = os.getcwd()
```

```python
file_name = 'Energy Indicators.xls'

energy = pd.read_excel(file_name, skiprows = 16, usecols = "C:F")
energy = energy[1:-38]
energy.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
energy = energy.replace('...', np.NaN)
energy['Energy Supply'] = energy['Energy Supply'] * 1000000
energy['Country'] = energy['Country'].str.replace(r'\d+$','',regex=True) 
energy['Country'] = energy['Country'].str.replace(r'\(.*\)', '', regex=True)
energy['Country'] = energy['Country'].str.strip()
energy['Country'] = energy['Country'].replace({'Republic of Korea': 'South Korea',
                'United States of America':'United States',
                'United Kingdom of Great Britain and Northern Ireland':'United Kingdom',
                'China, Hong Kong Special Administrative Region':'Hong Kong'})
energy
```

```python
file_name = 'gdp_data.csv'
GDP = pd.read_csv(file_name, sep = ',', header= None,
                 dtype = None, skiprows=4, encoding='utf-8')
GDP.columns = GDP.iloc[0]
GDP = GDP[1:]
GDP.columns = [int(col) if isinstance(col, float) and col.is_integer() else col for col in GDP.columns]
GDP['Country Name'] = GDP['Country Name'].replace({'Korea, Rep.': 'South Korea',
                'Iran, Islamic Rep.':'Iran',
                'Hong Kong SAR, China':'Hong Kong'})
GDP.rename(columns={'Country Name' : 'Country'}, inplace=True)

GDP
```

```python
file_name = 'scimagojr.xlsx'
ScimEn = pd.read_excel(file_name)
ScimEn
```

```python
United = pd.merge(ScimEn, energy, how='inner', on='Country')
United = pd.merge(United, GDP, how='inner', on='Country')
United = United[:15]
United = United.drop(['Region', 'Country Code', 'Indicator Name', 'Indicator Code'], axis=1)
United = United.drop(United.columns[-1], axis=1)
United = United.drop([col for col in United.columns 
                      if isinstance(col, (int, float)) and (col < 2006 or col > 2015)], axis=1)
United = United.set_index('Country')

United
```

```python
def answer_two():
    avgGDP = pd.Series()
    years = list(range(2006, 2016)) 
    for idx, row in United.iterrows():            
        avgGDP.loc[idx] = row[years].mean()
    avgGDP = avgGDP.sort_values(ascending=False)
    return avgGDP

print(answer_two())
    
```

```python
def answer_three():
    country = answer_two().index[5]
    return int(United.loc[country, 2015] - United.loc[country, 2006])

print(answer_three())
```

```python
def answer_four():
    United['Ratio'] = United['Self-citations'] / United['Citations']

    tuple = (United['Ratio'].idxmax(), United['Ratio'].max()) 
    return tuple

print(answer_four())
     
```

```python
def answer_five():
    United['Population'] = United['Energy Supply'] / United['Energy Supply per Capita']
    return United['Population'].nlargest(3).idxmin()

print(answer_five())
```

```python
def answer_six():
    United['Citable documents per Capita'] = United['Citable documents'] / United['Population']
    return United['Citable documents per Capita'].corr(United['Energy Supply per Capita'])


print(answer_six())
```

```python
ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}

def answer_seven():
    Continents = pd.DataFrame({'Continent' : ['Asia', 'Australia', 'Europe', 'North America', 'South America']})
    Continents = Continents.set_index('Continent')
    Continents['Size'] = 0
    Continents['Sum'] = 0
    continents_data = {
    'Asia': [],
    'Australia': [],
    'Europe': [],
    'North America': [],
    'South America': []
}
    for country, continent in ContinentDict.items():
        Continents.loc[continent, 'Size'] += 1
        Continents.loc[continent, 'Sum'] += United.loc[country, 'Population']
        continents_data[continent].append(United.loc[country, 'Population'])
    Continents['Mean'] = Continents['Sum'] / Continents['Size']
    std_devs = {cont: np.std(pops, ddof=1) for cont, pops in continents_data.items()}
    Continents['Std deviation'] = Continents.index.map(std_devs)
    return Continents

    

a = answer_seven()
a
```
